```python
import pandas as pd
import gzip as gzip
import ast
import json
```

```python
archivo = "../dataSet/steam_games.json.gz"

data = []

with gzip.open(archivo, 'rt', encoding='utf-8') as file:
    for line in file:
        try:
            cleaned_line = line.replace('NaN', 'null')
            json_data = json.loads(cleaned_line)
            data.append(json_data)
        except json.JSONDecodeError as e:
            print(f"Error en la l√≠nea: {line}")
            print(f"Detalle del error: {e}")
            continue

df = pd.DataFrame(data)
```

```python
df_steam_games = df.dropna(how='all')
```

```python
#identificamos los valores unicos de la columna price
df_steam_games['price'].unique()
```

```python
df_steam_games['price'] = df_steam_games['price'].replace({ 'Starting at $499.00': 499.0})
df_steam_games['price'] = df_steam_games['price'].replace({ 'Starting at $449.00': 449.0})
```

```python
# Convertir valores no num√©ricos a NaN
df_steam_games['price'] = pd.to_numeric(df_steam_games['price'], errors='coerce')

# Reemplazar NaN con 0.0
df_steam_games['price'].fillna(0.0, inplace=True)

# Verificar los cambios
df_steam_games.head()
```

```python
df_steam_games['price'].dtype
```

```python
#1. identificar los registros duplicados que tiene la columna id
duplicated_ids = df_steam_games[df_steam_games.duplicated('id', keep=False)]

#2. #1. mostrar los registros duplicados que tiene la columna id
duplicated_ids
```

```python
# Eliminar duplicados, manteniendo solo la primera aparici√≥n en la columna id

#Crear m√°scara booleana para excluir el primer duplicado (NaN)
mascara_bool = ~df_steam_games.index.isin([duplicated_ids])

#Seleccionar filas con la m√°scara
df_new = df_steam_games[mascara_bool]

#Reemplazar df_steam_games con solo valores verdaderos de la m√°scara
df_steam_games = df_new.copy()

#se corrobora que se elimina el registro
filtered_rows = df_steam_games[df_steam_games['id'].isna()]
filtered_rows

```

```python
df_steam_games['release_date'].isna().value_counts()
```

```python
# Eliminar las filas donde 'release_date' es None o NaN
df_steam_games = df_steam_games.dropna(subset=['release_date'])
```

```python
df_steam_games['release_date'].isna().value_counts()
```

```python
# Convertir la columna 'release_date' a datetime, convirtiendo errores a NaT
df_steam_games['release_date'] = pd.to_datetime(df_steam_games['release_date'], errors='coerce')
```

```python
#Eliminar filas con NaT en 'release_date'
df_steam_games_cleaned = df_steam_games.dropna(subset=['release_date'])
```

```python
# Asegurarse de que 'release_date' est√° en formato datetime
df_steam_games['release_date'] = pd.to_datetime(df_steam_games['release_date'])
df_steam_games['Year'] = df_steam_games['release_date'].dt.year
# Eliminar la columna 'release_date'
df_steam_games = df_steam_games.drop(columns=['release_date'])
```

```python
df_steam_games.head()
```

```python
# Columnas que necesitan cambiar los valores nulos
columns_to_fill = ['publisher', 'app_name', 'title', 'developer']
# Reemplazar valores nulos con 'Available'
df_steam_games[columns_to_fill] = df_steam_games[columns_to_fill].fillna('Available')
```

```python
df_steam_games
```

```python
df_steam_games = df_steam_games.explode('genres')
df_steam_games = df_steam_games.dropna(subset=['genres'])
df_steam_games.head()
```

```python
#se eliminan las columnas tags, specs, url y reviews_url al conciderarse que no son necesarias
df_steam_games = df_steam_games.drop(['tags', 'specs', 'url', 'reviews_url'], axis=1)

df_steam_games.columns
```
